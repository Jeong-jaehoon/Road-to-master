### 4.1 ÏÑ†ÌòïÌöåÍ∑Ä

#### 4.1.1 Ï†ïÍ∑úÎ∞©Ï†ïÏãù

```python
import numpy as np
X = 2 * np.random.rand(100, 1) #0 Í≥º 100ÏÇ¨Ïù¥Ïóê Î¨¥ÏûëÏúÑÎ°ú ÎÇúÏàò Ï∂îÏ∂úÌõÑ X 2
y = 4 + 3 * X + np.random.randn(100, 1) # 4+ 3x + bias ÌòïÌÉú biasÎäî ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÏÉòÌîå 100
```

```python
X_b = np.c_[np.ones((100, 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
import matplotlib.pyplot as plt

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0,2,0,15])
plt.show()
```

#### 4.2.1 Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï

Ìé∏ÎØ∏Î∂ÑÏùÑ Ïù¥Ïö©Ìï¥ÏÑú Î™®Îì† ÌååÎùºÎØ∏ÌÑ∞Ïóê ÎåÄÌï¥ Ìé∏ÎèÑÌï®ÏàòÎ•º Íµ¨ÌïúÎã§.

Ïù¥ Î∞©Î≤ïÏùÄ Îß§ Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï Ïä§ÌÖùÏóêÏÑú Ï†ÑÏ≤¥ ÌõàÎ†® ÏÑ∏Ìä∏ XÏóê ÎåÄÌï¥ÏÑú Í≥ÑÏÇ∞ÌïúÎã§.  
Í∑∏ÎûòÏÑú **Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï**Ïù¥ÎùºÍ≥† ÌïúÎã§.  
Ï¶â, Îß§ Ïä§ÌÖùÏóêÏÑú ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤¥Î•º ÏÇ¨Ïö©ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Îß§Ïö∞ ÌÅ∞ ÌõàÎ†®ÏÑ∏Ìä∏ÏóêÏÑúÎäî ÏïÑÏ£º ÎäêÎ¶¨Îã§.  
ÎåÄÏã†Ïóê Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÄ ÌäπÏÑ±ÏàòÏóê ÎØºÍ∞êÌïòÏßÄ ÏïäÍ∏∞ ÎïåÎ¨∏Ïóê  
ÏàòÏã≠ÎßåÍ∞úÏùò ÌäπÏÑ±ÏóêÏÑú ÏÑ†ÌòïÌöåÍ∑ÄÎ•º ÌõàÎ†®ÏãúÌÇ§Î†§Î©¥ Ï†ïÍ∑úÎ∞©Ï†ïÏãùÎ≥¥Îã§ Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏù¥ Îπ†Î•¥Îã§.

*Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùò Ïä§ÌÖù*

$\theta^{next step} = \theta - \eta  d\theta MSE(\theta)$

*Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùò Íµ¨ÌòÑ*

```python
eta = 0.1 # ÌïôÏäµÎ•† (learning rate)
n_iterations = 1000
m = 100

theta = np.random.randn(2,1) # Î¨¥ÏûëÏúÑ Ï¥àÍ∏∞Ìôî

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏúºÎ°ú Ï∞æÏùÄ thetaÏôÄ Ï†ïÍ∑úÎ∞©Ï†ïÏãùÏúºÎ°ú Ï∞æÏùÄ theta_bestÍ∞íÏù¥ Ï†ïÌôïÌûà ÏùºÏπò!

**ÌïôÏäµÎ•†Ïóê Îî∞Î•∏ ÏÑ†ÌòïÌöåÍ∑ÄÏóêÏÑúÏùò gradient descent**

```python
theta_path_bgd = []

def plot_gradient_descent(theta, eta, theta_path = None):
    m = len(X_b)
    plt.plot(X, y, "b.")
    n_iterations = 1000
    for iteration in range(n_iterations):
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            style = "b-" if iteration >0 else "r--"
            plt.plot(X_new, y_predict, style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradients
        if theta_path is not None:
            theta_path.append(theta)
    plt.xlabel("$x_1$", fontsize = 18)
    plt.axis([0,2,0,15])
    plt.title(r"$\eta = {}$".format(eta), fontsize = 18)
```

```python
np.random.seed(42)
theta = np.random.randn(2,1) # random initialization

plt.figure(figsize = (10,4))
plt.subplot(131); plot_gradient_descent(theta, eta = 0.02)
plt.ylabel("$y$", rotation =0, fontsize = 18)
plt.subplot(132); plot_gradient_descent(theta, eta = 0.1, theta_path = theta_path_bgd)
plt.subplot(133); plot_gradient_descent(theta, eta = 0.5)

plt.show()
```

#### 4.2.2 ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï


Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÄ Îß§ Ïä§ÌÖùÏóêÏÑú Ï†ÑÏ≤¥ ÌõàÎ†® ÏÑ∏Ìä∏Î•º ÏÇ¨Ïö©Ìï¥ Í∑∏ÎûòÎîîÏñ∏Ìä∏Î•¥ Í≥ÑÏãºÌï¥ ÌõàÎ†®ÏÑ∏Ìä∏Í∞Ä Ïª§ÏßÄÎ©¥ Îß§Ïö∞ ÎäêÎ†§ÏßÑÎã§  
Ïù¥ÏôÄÎäî Î∞òÎåÄÎ°ú **ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï**ÏùÄ Îß§ Ïä§ÌÖùÏóêÏÑú Ìïú Í∞úÏùò ÏÉòÌîåÏùÑ Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌïòÍ≥† Í∑∏ ÌïòÎÇòÏùò ÏÉòÌîåÏóê ÎåÄÌïú Í∑∏ÎûòÎîîÏñ∏Ìä∏Î•º Í≥ÑÏÇ∞ÌïúÎã§. Ï†ÅÏùÄ Îç∞Ïù¥ÌÑ∞Îßå Ï≤òÎ¶¨ÌïòÍ∏∞ÎïåÎ¨∏Ïóê ÏïåÍ≥†Î¶¨Ï¶òÏù¥ Ìõ®Ïî¨ Îπ†Î•¥Í≥† ÌÅ∞ Îç∞Ïù¥ÌÑ∞Î•º Îã§Î£∞ Ïàò ÏûàÎã§.

Î∞òÎ©¥ ÌôïÎ•†Ï†ÅÏù¥Í∏∞ÎïåÎ¨∏Ïóê ÏïåÍ≥†Î¶¨Ï¶òÏù¥ Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÎ≥¥Îã§ Ìõ®Ïî¨ Î∂àÏïàÏ†ïÌïòÎã§.  
ÎπÑÏö© Ìï®ÏàòÍ∞Ä ÏµúÏÜüÍ∞íÏóê Îã§Îã§Î•º ÎïåÍπåÏßÄ Î∂ÄÎìúÎüΩÍ≤å Í∞êÏÜåÌïòÏßÄÏïäÍ≥† ÏöîÎèôÏπ®.

Îã®, ÎπÑÏö©Ìï®ÏàòÍ∞Ä Î∂àÍ∑úÏπôÏ†ÅÏùºÍ≤®Ïö∞ ÏïåÍ≥†Î¶¨Ï¶òÏù¥ ÏßÄÏó≠ ÏµúÏÜüÍ∞íÏùÑ Í±¥ÎÑà Îõ∏Ïàò ÏûàÎèÑÎ°ù ÎèÑÏôÄÏ£ºÎØÄÎ°ú  
ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏù¥ Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÎ≥¥Îã§ Ï†ÑÏó≠ ÏµúÏÜüÍ∞íÏùÑ Ï∞æÏùÑ Í∞ÄÎä•ÏÑ±Ïù¥ ÎÜíÎã§

Î¨¥ÏûëÏúÑÏÑ±ÏùÄ ÏßÄÏó≠ ÏµúÏÜüÍ∞íÏóêÏÑú ÌÉàÏ∂úÏãúÏºúÏ§òÏÑú Ï¢ãÏßÄÎßå ÏïåÍ≥†Î¶¨Ï¶òÏùÑ Ï†ÑÏó≠ ÏµúÏÜüÍ∞íÏóê Îã§Îã§Î•¥ÏßÄ Î™ªÌïòÍ≤å ÌïúÎã§Îäî Ï†êÏóêÏÑú Ï¢ãÏßÄ ÏïäÎã§.  
Ïù¥ Ìï¥Í≤∞Î≤ïÏùÄ ÌïôÏäµÎ•†ÏùÑ Ï†êÏßÑÏ†ÅÏúºÎ°ú Í∞êÏÜåÏãúÌÇ§Îäî Î∞©Î≤ïÏù¥ÏûàÎã§.  
Ï¶â, ÏãúÏûëÌï†ÎñÑÎäî ÌïôÏäµÎ•†ÏùÑ ÌÅ¨Í≤åÌïòÍ≥† Ï†êÏ∞® ÏûëÍ≤å Ï§ÑÏó¨ÏÑú ÏïåÍ≥†Î¶¨Ï¶òÏù¥ Ï†ÑÏó≠ ÏµúÏÜüÍ∞íÏóê ÎèÑÎã¨ÌïòÍ≤å ÌïòÎäîÍ≤ÉÏù¥Îã§.  Ïù¥Î•º **ÌïôÏäµÏä§ÏºÄÏ§Ñ** Ïù¥ÎùºÍ≥† Î∂ÄÎ•∏Îã§.

#### ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï Íµ¨ÌòÑ

ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùò Ï≤´ 20Í∞ú step

```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
```

```python
n_epochs = 50
t0, t1 = 5, 50  # ÌïôÏäµ Ïä§ÏºÄÏ§Ñ ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # Î¨¥ÏûëÏúÑ Ï¥àÍ∏∞Ìôî

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    
            y_predict = X_new_b.dot(theta)           
            style = "b-" if i > 0 else "r--"         
            plt.plot(X_new, y_predict, style)        
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 

plt.plot(X, y, "b.")                                 
plt.xlabel("$x_1$", fontsize=18)                     
plt.ylabel("$y$", rotation=0, fontsize=18)           
plt.axis([0, 2, 0, 15])                                       
plt.show()               
```

#### 4.2.3 ÎØ∏ÎãàÎ∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï

ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÍ≥º Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï Ï§ëÍ∞Ñ Ïñ¥ÎîòÍ∞ÄÏóê ÏûàÎäî ÎØ∏ÎãàÎ∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï,  
Ï¶â, ÎØ∏ÎãàÎ∞∞ÏπòÎùº Î∂ÄÎ•¥Îäî ÏûÑÏùòÏùò ÏûëÏùÄ ÏÉòÌîåÏÑ∏Ìä∏Ïóê ÎåÄÌï¥ Í∑∏ÎûòÎîîÏñ∏Ìä∏Î•º Í≥ÑÏÇ∞Ìï®.  
ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏóê ÎπÑÌï¥ ÎØ∏ÎãàÎ∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùò Ï£ºÏöî Ïû•Ï†êÏùÄ ÌñâÎ†¨ Ïó∞ÏÇ∞Ïóê ÏµúÏ†ÅÌôîÎêú ÌïòÎìúÏõ®Ïñ¥ Ï¶â GPUÏóêÏÑú ÏñªÎäî ÏÑ±Îä• Ìñ•ÏÉÅÏù¥Îã§.

ÎØ∏ÎãàÎ∞∞Ïπò ÌÅ¨Í∏∞Î•º Ïñ¥ÎäêÏ†ïÎèÑ ÌÅ¨Í≤åÌïòÎ©¥ ÌååÎùºÎØ∏ÌÑ∞ Í≥µÍ∞ÑÏóêÏÑú SGDÎ≥¥Îã§ Îçú Î∂àÍ∑úÏπôÌïòÍ≤å ÏõÄÏßÅÏù∏Îã§. ÌïòÏßÄÎßå ÏßÄÏó≠ ÏµúÏÜüÍ∞íÏóê Îπ†ÏßÄÍ≤å Îê† ÏóºÎ†§Îäî SGDÎ≥¥Îã§ ÎçîÌÅ¨Îã§.

```python
theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # Î¨¥ÏûëÏúÑ Ï¥àÍ∏∞Ìôî

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
```

ÌïôÏäµ Í≥°ÏÑ†ÏùÄ ÌõàÎ†®ÏÑ∏Ìä∏ÏôÄ Í≤ÄÏ¶ùÏÑ∏Ìä∏Ïùò Î™®Îç∏ ÏÑ±Îä•ÏùÑ ÌõàÎ†® ÏÑ∏Ìä∏ ÌÅ¨Í∏∞Ïùò Ìï®ÏàòÎ°ú ÎÇòÌÉÄÎÇ∏Îã§. Ïù¥ Í∑∏ÎûòÌîÑÎ•º ÏÉùÏÑ±ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Îã®ÏàúÌûà Ìõà"Î†® ÏÑ∏Ìä∏ÏóêÏÑú ÌÅ¨Í∏∞Í∞Ä Îã§Î•∏ ÏÑúÎ∏å ÏÑ∏Ìä∏Î•º ÎßåÎì§Ïñ¥ Î™®Îç∏ÏùÑ Ïó¨Îü¨ Î≤à ÌõàÎ†®ÏãúÌÇ§Î©¥ ÎêúÎã§. 

```python

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="ÌõàÎ†®")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Í≤ÄÏ¶ù")
    plt.legend(loc="upper right", fontsize=14)   # Ï±ÖÏóêÎäî Îπ†Ïßê
    plt.xlabel("ÌõàÎ†® ÏÑ∏Ìä∏ ÌÅ¨Í∏∞", fontsize=14)       # Ï±ÖÏóêÎäî Îπ†Ïßê
    plt.ylabel("RMSE", fontsize=14)              # Ï±ÖÏóêÎäî Îπ†Ïßê
```

Îã®ÏàúÏÑ†ÌòïÌöåÍ∑ÄÏóêÏÑúÏùò ÌïôÏäµ Í≥°ÏÑ†

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.axis([0, 80, 0, 3])                         # Ï±ÖÏóêÎäî Îπ†Ïßê
plt.show()       
```

```python
from sklearn.linear_model import Ridge

np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)

def plot_model(model_class, polynomial, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=3)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])

plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)

plt.show()
```

#### 4.5.4 Ï°∞Í∏∞ Ï¢ÖÎ£å

Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÍ≥º Í∞ôÏùÄ Î∞òÎ≥µ ÌïôÏäµ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ Í∑úÏ†úÌïòÎäî Îã§Î•∏ Î∞©Î≤ÖÏùÄ Í≤ÄÏ¶ùÏóêÎü¨Í∞Ä ÏµúÏÜüÍ∞íÏóê ÎèÑÎã¨ÌïòÎ©¥ Î∞îÎ°ú ÌõàÎ†®ÏùÑ Ï§ëÏßÄ ÏãúÌÇ§Îäî Í≤ÉÏù¥Îã§. Ïù¥Î•º __Ï°∞Í∏∞Ï¢ÖÎ£å__ ÎùºÍ≥†ÌïúÎã§

*ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏù¥ÎÇò ÎØ∏ÎãàÎ∞∞Ïπò Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÄ Í≥°ÏÑ†Ïù¥ Í∑∏Î¶¨ Îß§ÎÅÑÎüΩÏßÄ ÏïäÏïÑ ÏµúÏÜüÍ∞íÏóê ÎèÑÎã¨ÌñàÎäîÎç∞ ÌôïÏù∏ÌïòÍ∏∞ Ïñ¥Î†§Ïö∏ Ïàò ÏûàÎã§. Ïù¥Îïå Í≤ÄÏ¶ù ÏóêÎü¨Í∞Ä ÏùºÏ†ï ÏãúÍ∞ÑÎèôÏïà ÏµúÏÜüÍ∞íÎ≥¥Îã§ ÌÅ¥Îïå ÌïôÏäµÏùÑ Î©àÏ∂îÍ≥† Í≤ÄÏ¶ù ÏóêÎü¨Í∞Ä ÏµúÏÜåÏòÄÏùÑÎïåÏùò Î™®Îç∏ ÌååÎùºÎØ∏ÌÑ∞Î°ú ÎèåÎ¶¨ÎäîÍ≤ÉÏù¥ ÌïúÍ∞ÄÏßÄ Î∞©Î≤ïÏù¥Îã§.*

```python

import warnings
warnings.filterwarnings(action = 'ignore')
#Í≤ΩÍ≥†Î¨¥Ïãú
#####
from sklearn.base import clone


np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m, 1)

X_train, X_val, y_train, y_val = train_test_split(X[:50], y[:50].ravel(), test_size=0.5, random_state=10)

poly_scaler = Pipeline([
        ("poly_features", PolynomialFeatures(degree=90, include_bias=False)),
        ("std_scaler", StandardScaler()),
    ])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)

sgd_reg = SGDRegressor(max_iter=1,
                       penalty=None,
                       eta0=0.0005,
                       warm_start=True,
                       learning_rate="constant",
                       random_state=42)

n_epochs = 500
train_errors, val_errors = [], []
for epoch in range(n_epochs):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_train_predict = sgd_reg.predict(X_train_poly_scaled)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    train_errors.append(mean_squared_error(y_train, y_train_predict))
    val_errors.append(mean_squared_error(y_val, y_val_predict))

best_epoch = np.argmin(val_errors)
best_val_rmse = np.sqrt(val_errors[best_epoch])

plt.annotate('ÏµúÏÑ†Ïùò Î™®Îç∏',
             xy=(best_epoch, best_val_rmse),
             xytext=(best_epoch, best_val_rmse + 1),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=16,
            )

best_val_rmse -= 0.03  # just to make the graph look better
plt.plot([0, n_epochs], [best_val_rmse, best_val_rmse], "k:", linewidth=2)
plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Í≤ÄÏ¶ù ÏÑ∏Ìä∏")
plt.plot(np.sqrt(train_errors), "r--", linewidth=2, label="ÌõàÎ†® ÏÑ∏Ìä∏")
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("ÏóêÌè¨ÌÅ¨", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.show()
```
