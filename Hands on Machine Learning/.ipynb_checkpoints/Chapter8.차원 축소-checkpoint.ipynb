# 8장 차원 축소 

많은 경우 머신러닝은 훈련 샘플이 수천~수백만개의 특성을 가질 수도 있다. 이는 훈련을 느리게 할 뿐만 아니라 좋은 솔루션을 찾기 어렵게 한다. 이를 `차원의 저주` 라고한다.

`Caution`  
차원을 축소시키면 일부 정보가 유실된다.  
그래서 훈련 속도가 빨라질수는 있지만 시스템의 성능이 감소할 우려가 있다.  
또한 작업 파이프라인이 복잡해지고 유지 관리 또한 어려워진다. 그러므로 차워 축소를 고려하기전에 원본데이터로 시스템을 훈련시켜봐야한다.  
예외 적인 경우로 훈련데이터의 차원을 축소해서 잡음이나 불필요한 세부사항을 걸러내므로 성능을 높일 수 도 있다.(대부분의 경우는 훈련 속도만 빨라진다.)

## 8.1 차원의 저주

차원이 커질수록 데이터의 밀도가 낮아진다.  
다시 말하면 차원(dimension ~ column)이 클수록 데이터(row)가 기하급수적으로 많이 필요하다는 뜻이다.  
특성이 단 100개 라고해도 특정조건에서 우주의 모든 원자 수를 다합친것보다 큰 훈련샘플이 필요하다.  
즉 고차원 데이터는 일반적인 방법으로 모델링하기 상당히 힘들다 그래서 차원축소가 필요하다..

### 8.2.2 매니폴드 학습

스위스 롤은 2D 매니폴드의 한 예.  
2D매니폴드는 고차원 공간에서 휘어지거나 뒤틀린 2D모양이다.  
국부적으로는 2차원 평면으로 보이지만 실제로는 3차원 구조를 가지고있다.  

많은 차원 축소 알고리즘이 훈련 샘플에 놓여있는 매니폴드를 모델링 하는식으로 작동한다.  
이는 대부분 실제 고차원 데이터셋이 더 낮은 저차원 매니폴드에 가깝게 놓여있다는 __매니폴드 가정__ 에 근거한다. 이는 경험적으로 상당히 많이 발견된다.


##### 요약하면, 모델을 훈련시키기 전에 훈련 세트의 차원을 감소시키면 훈련 속도는 빨라지지만 항상 더 낫거나 간단한 솔루션이 되지는 않는다는 것. 이는 전적으로 데이터셋에 달려있다.

## 8.3 PCA

__주성분 분석(PCA)__ 는 가장 인기 있는 차원 축소 알고리즘이다. 먼저 데이터에 가장 가까운 초평면을 정의한다음 데이터를 이 평면에 투영시킨다.  


### 8.3.1 분산 보존

저차원의 초평면에 훈련 세트를 투영하기전에 먼저 올바른 초평면을 선택해야한다.  
다른방향으로 투영하는 것보다 분산이 최대로 보존되는 축을 선택하는 것이 정보가 가장 적게 손실되므로 가장 합리적으로 보인다. 이는 원본 데이터 세소가 투영된 것 사이의 평균 제곱 거리를 최소화 하는 축이된다.

### 8.3.2 주성분

첫번째 주성분 축은 분산이 최대인 축이 되고, 두번째 부터는 첫번째 축에 직교하고 남은 분산을 최대한 보존하는 축이된다.  
i번째 축을 정의하는 단위 벡터를 i번째 주성분(PC)라고 부른다.  

`NOTE_`  
주성분의 방향은 일정하지 않다 훈련 세트를 조금 섞은 다음 다시 PCA를 적용하면 새로운 PC중 일부가 원래 PC와 반대 방향일 수 있다. 어떤 경우에는 PC가 회전 하거나 바뀔 수 있지만 보통은 같은 평면을 구성한다.

PCA는 평균이 0이라고 가정하고 수행된다. 사이킷런의 PCA는 이 작업을 대신 해준다. 하지만 만약 직접 수행해야한다면 데이터를 원점에 맞춰주는걸 잊으면 안된다.

```python
X_centered = X - X.mean(axis = 0)
U, s, Vt = np.linalg.svd(X_centered)

c1 = Vt.T[:,0]
c2 = Vt.T[:,1]
```

$ X_{d-proj} = X W_d$

### 8.3.6 적절한 차원수 선택하기

축소할 차원수는 임의로 정하는 것보다 충분한 분산이 될 때 까지 더해야 할 차원 수를 선택하는 쪽을 더 선호한다.  

```python
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) +1
```

### 8.3.7 압축을 위한 PCA

차원을 축소하고나면 훈련 세트의 크기가 줄어든다. 예를들어 MNIST 데이터셋에 분산의 95%를 유지하도록 PCA를 적용해보자. 그러면 원래의 784개의 특성이 아닌 150개 정도만 가지게 된다.  
즉 대부분의 분산은 줄어들었지만 원본 데이터셋은 20%가량으로 줄어들었다.  
또한 PCA로 압축했던 데이터를 다시 재구성 할수있다.  
재구성된 데이터와 원본 데이터 사이의 평균 제곱 거리를 __재구성 오차(reconstruction error)__ 이라고한다.  

```python
pca = PCA(n_components = 154)
X_reduced = pca.fit_transform(X_train)
X_recovered = pca.inverse_transform(X_reduced)
```

$X_{recovered} = X_{d-proj}W^T_d$

### 8.3.8 점진적 PCA

PCA 구현의 문제점은 SVD 알고리즘을 실행하기 위해 전체 훈련 세틀르 메모리에 올려야 한다는 것.  
이를 해결하기위해 __점진적 PCA(Incremental PCA)__ 가 개발되었다.  
훈련 세트를 미니배치로 나눈뒤 IPCA 알고리즘에 한 번에 하나씩 주입한다. 이런 방식은 훈련 세트가 클때 유용하고 온라인으로 PCA를 적용할 수도있다.  

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X_train, n_batches):
    inc_pca.partial_fit(X_batch)
    
X_reduced = inc_pca.transform(X_train)
```

### 8.3.9 랜덤 PCA

사이킷런에는 PCA의 또 다른 옵션으로 __랜덤 PCA__를 제공한다.  
이는 확률적인 알고리즘으로 첫 d개의 주성분에 대한 근삿값을 빠르게 찾는다. (속도적 향상)  

```python
rnd_pca = PCA(n_components= 154, svd_solver='randomized')
X_reduced = rnd_pca.fit_transform(X_train)
```
