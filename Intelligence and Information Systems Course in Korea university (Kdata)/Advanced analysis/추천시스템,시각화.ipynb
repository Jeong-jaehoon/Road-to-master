```python
def recommendation_user(user_id, k, n):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index
    temp = pd.DataFrame(columns =['user_id','item_id','rating'])
    for j, i in enumerate(s, 1):
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id==user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id == i)] ##서로안봤던 영화중에서 rating이 5고
        k.rating = k.rating * (1-0.05*j)
        temp = pd.concat([temp, k])
    else:
        return temp.groupby('item_id').mean().sort_values('rating', ascending=False)[:n]
```

### 선생님 코드

```python
def recommendation_user(user_id, k,n=10):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    for j, i in enumerate(s,1):
        t = set(data[data.user_id==i].item_id)  - set(data[data.user_id==user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id==i)]
        k.rating = k.rating * (1-0.05*j)
        temp = pd.concat([temp, k])
    else:
        return temp.rename(columns={'item_id':'movie_id'}).merge(item[['movie_id','movie_title']]).drop_duplicates().groupby('movie_title').mean().sort_values('rating',ascending=False)[:n].index
```
