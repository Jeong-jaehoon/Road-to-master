```python
import json
import urllib.request

url = "http://openapi.airkorea.or.kr/openapi/services/rest/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty" 
##? 이전의 url 주소.
params = {
    "serviceKey" : "X83SzHZu2hN3oX5qNXX%2Fj1HMKIxz46x9gZC%2BULtBTZw83s53dP68PymkE%2FLNjukp0npKAMhb5CAIg%2BLlW8WedA%3D%3D",
    "numOfRows":10,
    "pageNo" : 1,
    "startPage":1,
    "stationName":"성북구",
    "dataTerm":"DAILY",
    "sidoName":"인천",
    "ver": "1.3",
    "_returnType":"JSON",
}

# 인코딩 과정
params["serviceKey"] = urllib.parse.unquote(params["serviceKey"]) ## parsing한 키를 넣어준다..?
print(type(params)) ## 이떄 type은  parmas 의 type은 여전히 dict
params = urllib.parse.urlencode(params) ## 인코딩할거를 parmas에 넣어준다.
print(type(params)) ## 이떄 type은 string
params = params.encode("utf-8") ## parmas을 "utf-8" 형식으로 encode
print(type(params)) ## 이제 bytes형식이 되었음.

# REQUEST 과정

req = urllib.request.Request(url, data=params) ##req에다가 ...?
print(type(req)) ##이떄 type은 urllib.request.Request
res = urllib.request.urlopen(req) ## res 에다가는 ...?
print(type(res)) ## 이때 type은 http.client.HTTPResponse'

# RESPONSE 과정 - DECODE
resStr = res.read() ## res를 Read해줌
print(type(resStr)) ##이떄 res는 bytes type
resStr = resStr.decode("utf-8") ## utf-8형식을 decode해줌.

resObj = json.loads(resStr)  ##resObj 에 resStr을 loads
print(type(resObj)) ##이떄 resObj의 형식은 dictionary

resJSON = json.dumps(resObj, ensure_ascii = False,  indent="  ")  ##resJSON 에 덤프 걸어준다 이때 ensure_ascii = False 옵션으로 한글 보이게함
print(type(resJSON))  ##이때 resJSON의 형식은 string

print(resJSON) 

resObj["list"][0]["pm10Grade"]
```

#### Tree 만들기

```python
bookStore = et.Element("bookstore")

book1 = et.Element("book", category="cooking")
bookStore.insert(0, book1)

title1 = et.Element("title")
title1.attrib["lang"] = "en"
title1.text = "Everyday Italian"
book1.append(title1)

et.SubElement(book1, "author").text = "GIada De Laurentiis"
et.SubElement(book1, "year").text = "2005"
et.SubElement(book1, "price").text = "30.00"

book2 = et.Element("book", {"category":"children"})
bookStore.append(book2)

title2 = et.Element("title")
title2.attrib["lang"] =title1.get("lang")
title2.text =  "Harry Potter"
book2.append(title2)

et.SubElement(book2, "author").text = "GIada De Laurentiis"
et.SubElement(book2, "year").text = "2005"
et.SubElement(book2, "price").text = "30.00"

et.dump(bookStore)

## 여기 나온결과를 html로 넣으면은? 된다.
```
