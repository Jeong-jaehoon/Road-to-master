#### 정리해보자면 판다스에서 row, column indexing의 방법은 다섯가지가 있다

1. key방식  
    -  가장기본적인 방식이다
2. attribute 방식
    - 편하게 쓰기위해서 판다스에서 지원했지만 사이즈나 공백이 똑같은 이름이 있을때 잘동작하지않음 추천하지않는다
3. fancy indexing  
    - 데이터 프레임으로가져온다 (good)
4. loc 방식
    - 이름을 맞춰서 가져오는 방식
5. iloc 방식
    - 인덱스를 맞춰서 가져오는 방식, 인덱스이기 때문에 마지막 꺼 안가져옴 예르들어 3:6 이면 3,4,5 행을 가져온다.

속도자체는 numpy에서 가져온 key방식이 가장 빠르다고한다

특정 행, 열 가져오기 - at

```python
tips.at[3,'time']
```
