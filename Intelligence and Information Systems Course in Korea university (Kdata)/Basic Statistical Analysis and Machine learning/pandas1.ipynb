열 뽑아오는 방식 세가지

1. key 방식
    * 가장 기본적인 방식

2. attribute 방식
    * 편하게 쓰기위해서 판다스에서 지원했지만 사이즈나 공백이나 똑같은 이름이 있을때 잘 안됨

3. fancy indexing
    * 데이터프레임으로 가져온다
    
4. loc 방식

    * 이름을 맞춰서 가져오는방식
    
5. iloc 방식
    * 인덱스를 맞춰서 가져오는 방식 인덱스이기 때문에 마지막 꺼 안가져옴 예를들어 3:6 이면 3,4,5, 행 가져옴
    
    
> 그러면 속도는 어느게 빠를까
>> 키방식이 제일빠름 왜냐하면 넘파이로부터 나온친구이기때문
